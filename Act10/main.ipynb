```python
from openai import OpenAI

# Connect to LM Studio local server
client = OpenAI(
    base_url="http://127.0.0.1:1015/v1",
    api_key="lm-studio"  # any string works
)
```

```python
# Send a simple prompt
response = client.chat.completions.create(
    model="qwen/qwen3-1.7b",  # name can be anything in LM Studio
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Explain what a microcontroller is."}
    ],
    temperature=0.7
)

# Print response
print(response.choices[0].message.content)
```
