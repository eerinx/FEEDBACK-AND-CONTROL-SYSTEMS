```python
import google.generativeai as genai

# 1. I-setup ang API Key (Kunin sa Google AI Studio)
genai.configure(api_key="AIzaSyADRbfnOwlhcsbMIB8Ah8i14sUD8O7WS4Q")

# 2. Piliin ang model
# Pwedeng 'gemini-1.5-pro' o 'gemini-1.5-flash'
model = genai.GenerativeModel('gemini-3-flash-preview')

# 3. Chat loop (Katulad ng ginawa mo sa main.ipynb)
chat = model.start_chat(history=[])

while True:
    query = input("Ask Gemini: ")
    if query.lower() in ["exit", "quit"]:
        break
        
    response = chat.send_message(query, stream=True)
    
    for chunk in response:
        if chunk.text:
            print(chunk.text, end="", flush=True)
    print("\n")
```
