```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl
```

```python
# Define the Transfer Function G(s) = (s + 1) / (s^4 + 4s^3 + 7s^2 + 6s + 3)
num = [1, 1]
den = [1, 4, 7, 6, 3]
sys = ctrl.TransferFunction(num, den)

# Step Response
t, y = ctrl.step_response(sys)

plt.figure(figsize=(12, 5))

# Plot Step Response
plt.subplot(1, 2, 1)
plt.plot(t, y, color='#007B8A', linewidth=2)
plt.title('Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True, linestyle='--', alpha=0.7)
```

```python
# Plot Pole-Zero Map
plt.subplot(1, 2, 2)
ctrl.pzmap(sys)
plt.title('Pole-Zero Map')
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
