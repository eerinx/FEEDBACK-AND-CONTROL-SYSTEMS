```python
import control
import matplotlib.pyplot as plt
import numpy as np

# Define Numerator and Denominator
# G(s) = (4s^4 + 11s^2 + 9) / (16s^6 + 16s^5 + 68s^4 + 40s^3 + 68s^2 + 12s)
num = [4, 0, 11, 0, 9]
den = [16, 16, 68, 40, 68, 12, 0]

# System Definition
sys = control.TransferFunction(num, den)

# Print Analysis
print(f"Poles: {sys.poles()}")
print(f"Zeros: {sys.zeros()}")

# Plots
t, y = control.step_response(sys)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(t, y)
plt.title('Step Response (Marginally Stable Ramp-like Behavior)')
plt.xlabel('Time (s)')
plt.ylabel('x2(t)')
plt.grid(True)

plt.subplot(1, 2, 2)
control.pzmap(sys, plot=True)
plt.title('Pole-Zero Map')
plt.grid(True)

plt.tight_layout()
plt.show()
```
