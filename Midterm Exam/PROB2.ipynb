```python
import numpy as np
import matplotlib.pyplot as plt
import control as ct

# Define transfer function G(s) = 1 / (s * (s + 1) * (s**2 + s + 2))
num = [1]
den = [1, 2, 4, 2, 0] # Expand denominator to s^4 + 2s^3 + 4s^2 + 2s
sys = ct.TransferFunction(num, den)
```

```python
# Pole-Zero Visualization
plt.figure(figsize=(6, 6))
ct.pzmap(sys)
plt.title('Pole-Zero Map')
plt.grid(True)
plt.show()

# Output numeric results
print(f"Poles: {ct.poles(sys)}")
```

```python
# Step Response Analysis
t, y = ct.step_response(sys)
plt.figure(figsize=(10, 6))
plt.plot(t, y, color='#007B8A', linewidth=2)
plt.title('Step Response (Theta 2)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True, alpha=0.3)
plt.show()
```
